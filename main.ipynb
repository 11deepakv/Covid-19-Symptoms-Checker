# Handling Missing data

# Train Test Splitting

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(data_pre, data_pre['pneumonia']):
    strat_train_set = data_pre.iloc[train_index]
    strat_test_set = data_pre.iloc[test_index]
```

```python
data_prep = strat_train_set.copy()
```

```python
data_feature = strat_train_set.drop(['SARS-CoV-2 Positive'], axis=1)
data_label = strat_train_set['SARS-CoV-2 Positive'].copy()
```

```python
data_feature
```

```python
data_label.describe()
```

# Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('standard_scalr', StandardScaler())
])
```

```python
preprocess_data = my_pipeline.fit_transform(data_feature)
preprocess_data.shape
```

# Selecting the best model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import LinearRegression

model = LogisticRegression()
# model = DecisionTreeClassifier()
# model = RandomForestClassifier()
# model = svm.SVC()
# model = LinearRegression()
model.fit(preprocess_data, data_label)
```

```python
rand_feat = preprocess_data[:10]
rand_label = data_label[:10]
predication = model.predict(rand_feat)
predication
```

```python
list(rand_label)
```

# Evaluating the Model

```python
from sklearn.metrics import mean_squared_error
final_pred = model.predict(preprocess_data)
mse = mean_squared_error(data_label, final_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

Using Cross Validation Technique

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, preprocess_data, data_label, scoring='neg_mean_squared_error', cv=10)
-scores
```

```python
rmse_scores = np.sqrt(-scores)
```

```python
def print_score(score):
    print('Scores: ', score)
    print('Mean: ', score.mean())
    print('STD: ', score.std())    
```

```python
print_score(rmse_scores)
```

# Testing the Model

```python
final_predications = model.predict_proba(preprocess_data)
for pred, actual in zip(final_predications, data_label):
    print(pred[1]*100," ",actual)
```
